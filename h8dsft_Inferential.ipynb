## Hypothesis Testing

```python
dki = df[df['Province']=='DKI Jakarta'][['Date','New Cases']].groupby('Date').sum().sample(200)
jatim = df[df['Province']=='Jawa Timur'][['Date','New Cases']].groupby('Date').sum().sample(200)
```

My hypothesis on this case:

H0: μ_dki = μ_jabar

H1: μ_dki != μ_jabar

```python
t_stat, p_val = stats.ttest_ind(dki,jatim)
print('P-value:',p_val[0])
print('t-statistics:',t_stat[0])
```

### In this case, we get that P-value < 0,05. Where 0,05 is the critical point that i set. So we can reject the Null Hypothesis (H0). That means there is some significant difference between the data in DKI Jakarta and Jawa Timur.
