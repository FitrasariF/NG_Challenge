Buatlah sebuah kode untuk mensimulasikan melempar koin sebanyak 50.000 kali dengan assign value random berupa 0 dan 1 untuk kepala atau ekor. Setiap kali koin dilempar, peluang untuk mendapatkan kepala atau ekor adalah 50%.

## Head 

```python
import random

# Count the number of Heads results
h = 0

# Create a list of all results
results = []

# loop through 50000 trials
trials = 50000
trial = 0
while trial < trials:
    trial = trial + 1
    # Flip coin
    result = ['H' if random.randint(0,1) == 0 else 'T']
    results.append(result)
    # If it's heads, add it to the count
    h = h + int(result == ['H'])
    
# What proportion of trials produced heads
print ("%.2f%%" % ((h/trials)*100))

# Show all the results
print (results)
```

## Tail

```python
import random

# Count the number of Heads results
t = 1

# Create a list of all results
results = []

# loop through 50000 trials
trials = 50000
trial = 0
while trial < trials:
    trial = trial + 1
    # Flip coin
    result = ['T' if random.randint(0,1) == 1 else 'H']
    results.append(result)
    # If it's tails, add it to the count
    t = t + int(result == ['T'])
    
# What proportion of trials produced tails
print ("%.2f%%" % ((t/trials)*100))

# Show all the results
print (results)
```
